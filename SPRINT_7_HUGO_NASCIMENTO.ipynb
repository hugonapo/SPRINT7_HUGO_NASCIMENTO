Esta primera función es como una interfaz de entrada del programa. Sirve para recolectar, dentro de una secuencia de preguntas, la información que se necesita de cada aspirante al vacante, para eso se utiliza la función input(). Para formatear el tipo de respuesta que buscamos y para luego evaluar cada candidato, tenemos que informar al sistema que tipos de datos pueden ser inseridos, por ejemplo, float para los años de experiencia pues pueden ser numeros decimales, int para los niveles de competencia y estudios ya que buscamos una respuesta con numeros interos. Al final se agrupa toda la informacion en una lista estructurada llamada candidato que representa cada uno de los postulantes al vacante.

```python
def obtenir_dades():
    nombre = input('Cual es tu nombre y apellidos?:')
    telefono = input('Cual es tu numero de telefono?:')

    experiencia = float(input('Cuantos años llevas en la empresa?:'))
    catalan = int(input('Cual es tu nivel de catalan?:'))
    excel = int(input('Cual es tu nivel de excel?:'))
    titulo = int(input('Cual es tu titulo academico?:'))

    candidato = [nombre, telefono, experiencia, catalan, excel, titulo]
    return candidato
```

Esta función funciona como un filtro de calidad del programa. Ella recibe la lista de datos recopilados (candidato) y verifica, con la ayudade if y else, que los valores numericos recibidos cumplan con los rangos permitidos en el enunciado (valores entre 0 y 3 para catalán, excel y titulación, y experiencia no negativa). Si encuentra algún dato fuera de rango, devuelve False para descartar el registro, si todo es correcto devuelve True.

```python
def validar_dades(candidato):
  if candidato[2] < 0:
    return False
  if candidato[3] < 0 or candidato[3] > 3:
    return False
  if candidato[4] < 0 or candidato[4] > 3:
    return False
  if candidato[5] < 0 or candidato[5] > 3:
    return False
  else:
    return True
```

Aquí se gestiona la lógica de negocio o baremo de la empresa. La función extrae los datos numéricos de las posiciones específicas de la lista del candidato. Aplica el multiplicador ponderado a los años de experiencia (experiencia * 1.5) tal como exige el departamento de RRHH y le suma los puntos directos obtenidos en idiomas, herramientas y estudios, devolviendo una puntuación final de tipo flotante.

```python
def calcular_puntuacion(candidato):
  puntuacion = (candidato[2] * 1.5) + candidato[3] + candidato[4] + candidato[5]
  return puntuacion
```

El objetivo de esta función es segmentar el talento de forma automática basándose en la nota final calculada. Utiliza una estructura de control condicional (if/elif/else) para evaluar los límites de puntuación establecidos ( ≥12 , entre 8 y 12, o  <8 ). Para optimizar el posterior informe final, se añade un índice numérico al inicio de cada etiqueta ("1. Prioritat alta", "2. Candidat adequat", etc.). Esto permitirá que el algoritmo de ordenación alfabética sitúe los perfiles prioritarios en la cima de forma natural.

```python
def classificar_candidat(puntuacion):
  if puntuacion >= 12:
    return '1. Prioritat alta'
  elif puntuacion >= 8:
    return '2. Candidat adequat'
  else:
    return '3. No prioritari'
```

Esta función se encarga de la salida de datos inmediata en pantalla. Cada vez que un candidato finaliza con éxito su inscripción, la función recibe sus datos personales, su nota y su etiqueta de prioridad para imprimir una ficha limpia, visual y estética en la consola. Sirve para confirmar en tiempo real que el candidato ha sido procesado correctamente por el sistema.

```python
def mostrar_individual(candidato, puntuacion, clasificacio):
    print("\n===========================================")
    print("--- RESUMEN DEL CANDIDATO ---")
    print("===========================================")
    print(f"Nombre: {candidato[0]}")
    print(f"Telefono: {candidato[1]}")
    print(f"Puntuación: {puntuacion}")
    print(f"Clasificació: {clasificacio}")
```

Es la encargada de generar el cierre definitivo del proceso de selección cuando se alcanza el cupo. Recibe la lista global con los 10 candidatos válidos acumulados. Utiliza la función avanzada sorted() combinada con una estructura lambda para ordenar la matriz bajo un doble criterio: primero de forma alfabética por su nivel de prioridad (x[3]), y en caso de empate, de forma descendente por su puntuación (-x[2]). Una vez ordenada, un bucle for recorre la lista e imprime el listado final ordenado con el formato compacto solicitado para RRHH.

```python
def mostrar_informe(llista_candidats):
    print("\n=============================================")
    print("---      INFORME FINAL DE SELECCIÓ        ---")
    print("=============================================")


    llista_ordenada = sorted(llista_candidats, key=lambda x: (x[3], -x[2]))


    for candidat in llista_ordenada:

        print(f"Candidat/a: {candidat[0]} | Tel: {candidat[1]} | Punts: {candidat[2]} | {candidat[3]}")

    print("=============================================")
```

Esta función actúa como el "orquestador" o motor principal del programa (Main). Inicializa una lista vacía para almacenar las contrataciones y controla el flujo del programa mediante un bucle while que se ejecutará indefinidamente mientras no se alcancen exactamente 10 registros válidos (len(llista_final) < 10). En cada iteración coordina secuencialmente las funciones anteriores: solicita datos, si la validación es exitosa calcula, clasifica, muestra la ficha individual y almacena el resultado. Al completarse el cupo, rompe el bucle e invoca al informe final.

```python
def main_seleccio():
    print("--- INICIANDO SISTEMA DE RRHH (MÁXIMO 10 CANDIDATOS) ---\n")

    llista_final = []

    while len(llista_final) < 10:
        print(f"\nInscripción del candidato número {len(llista_final) + 1}:")

        candidat = obtenir_dades()

        if validar_dades(candidat) == True:
            nota = calcular_puntuacion(candidat)
            etiqueta = classificar_candidat(nota)

            mostrar_individual(candidat, nota, etiqueta)

            llista_final.append([candidat[0], candidat[1], nota, etiqueta])
        else:
            print("\n[ERROR] Los datos de este candidato no son válidos. No se añade a la lista. Inténtalo de nuevo.")


    mostrar_informe(llista_final)

main_seleccio()
```

En esta celda definimos la estructura de datos inicial. Para cumplir con el requisito de no utilizar estructuras condicionales (if), implementamos "look-up tables" (tablas de consulta) mediante diccionarios de Python. TABLA_COMPORTAMIENTO es un diccionario de diccionarios que asocia cada código numérico con su factor, su segmento de marketing y su mensaje personalizado. TABLA_EDAD mapea directamente los rangos de edad con sus correspondientes índices multiplicadores.

```python
clients = [
    {"id_client": 1, "nom": "Laura Serra", "edat": 23, "grup_edat": "18-25", "codi_comportament": 4},
    {"id_client": 2, "nom": "Marc Vidal", "edat": 35, "grup_edat": "26-40", "codi_comportament": 3},
    {"id_client": 3, "nom": "Ana López", "edat": 29, "grup_edat": "26-40", "codi_comportament": 2},
    {"id_client": 4, "nom": "Joan Riera", "edat": 19, "grup_edat": "18-25", "codi_comportament": 2},
    {"id_client": 5, "nom": "Mónica Pérez", "edat": 47, "grup_edat": "41-60", "codi_comportament": 4},
    {"id_client": 6, "nom": "Luis García", "edat": 52, "grup_edat": "41-60", "codi_comportament": 1},
    {"id_client": 7, "nom": "Pilar Sánchez", "edat": 61, "grup_edat": "60+", "codi_comportament": 3},
    {"id_client": 8, "nom": "Eva Martín", "edat": 38, "grup_edat": "26-40", "codi_comportament": 0},
    {"id_client": 9, "nom": "Diego Romero", "edat": 24, "grup_edat": "18-25", "codi_comportament": 1},
    {"id_client": 10, "nom": "Núria Costa", "edat": 33, "grup_edat": "26-40", "codi_comportament": 4}
]

TABLA_COMPORTAMIENTO = {
    0: {"factor": 0.10, "segment": "Navega pero no añade", "missatge": "¡Novedades exclusivas te esperan en la web!"},
    1: {"factor": 0.25, "segment": "Añade pero no checkout", "missatge": "¿Necesitas ayuda? Tu carrito te está esperando."},
    2: {"factor": 0.50, "segment": "Checkout pero no paga", "missatge": "¡Envío gratis si completas tu pedido ahora!"},
    3: {"factor": 0.75, "segment": "Abandonado < 24h", "missatge": "¡No pierdas tus artículos! Usa el cupón DESCUENTO5."},
    4: {"factor": 0.90, "segment": "Abandonado > 24h", "missatge": "Última oportunidad: guardamos tu carrito 12 horas más."}
}

TABLA_EDAD = {
    "18-25": 1.20,
    "26-40": 1.00,
    "41-60": 0.85,
    "60+":   0.70
}
```

Esta función recibe el código de comportamiento de un cliente y accede directamente al subdiccionario correspondiente dentro de TABLA_COMPORTAMIENTO. Al no usar condicionales, la búsqueda es directa y eficiente ($O(1)$). Devuelve simultáneamente tres valores empaquetados en una tupla: el factor numérico, el nombre del segmento y el mensaje de retargeting asignado

```python
def obtenir_segment_i_factor(codi):
    info = TABLA_COMPORTAMIENTO[codi]
    return info["factor"], info["segment"], info["missatge"]
```

Su objetivo es extraer el multiplicador asociado a la edad del cliente. Utiliza el parámetro de texto grup_edat como clave para realizar una consulta instantánea en el diccionario TABLA_EDAD, retornando el factor decimal correspondiente sin requerir evaluaciones lógicas intermedias.

```python
def obtenir_factor_edat(grup_edat):
    return TABLA_EDAD[grup_edat]
```

Esta función puramente matemática aplica la fórmula de negocio definida por el departamento de CRM ($risc\_final = factor\_codi \times factor\_edat$). Recibe los dos coeficientes previamente extraídos y devuelve el riesgo final redondeado a tres decimales para garantizar la limpieza visual de los datos.

```python
def calcular_risc(factor_codi, factor_edat):
    return round(factor_codi * factor_edat, 3)
```

Es la función encargada de transformar y enriquecer nuestro dataset. Recorre mediante un bucle for la lista de clientes originales y, para cada uno, invoca coordinadamente a las funciones de cálculo de las celdas anteriores. Para respetar la inmutabilidad de los datos de origen, genera una copia de cada diccionario (.copy()), le añade los cinco nuevos campos de análisis de marketing y los consolida en una nueva lista optimizada.

```python
def processar_clients(llista_clients):
    clients_processats = []
    for c in llista_clients:
        f_codi, seg, msg = obtenir_segment_i_factor(c["codi_comportament"])
        f_edat = obtenir_factor_edat(c["grup_edat"])
        risc = calcular_risc(f_codi, f_edat)


        nou_client = c.copy()
        nou_client["factor_codi"] = f_codi
        nou_client["factor_edat"] = f_edat
        nou_client["risc_final"] = risc
        nou_client["segment"] = seg
        nou_client["missatge"] = msg

        clients_processats.append(nou_client)
    return clients_processats
```

Esta función simula la representación visual de un DataFrame utilizando exclusivamente herramientas nativas de Python. Mediante marcadores de posición alineados (:<N), establece un ancho fijo para cada columna, logrando pintar en la consola una tabla perfectamente estructurada, simétrica y legible para los analistas de negocio.

```python
def mostrar_diccionari(dades_processades):
    print(f"{'ID':<4} | {'NOMBRE':<15} | {'EDAD':<5} | {'GRUPO':<6} | {'CODI':<4} | {'F_CODI':<6} | {'F_EDAD':<6} | {'RIESGO':<6} | {'SEGMENTO'}")
    print("-" * 110)
    for c in dades_processades:
        print(f"{c['id_client']:<4} | {c['nom']:<15} | {c['edat']:<5} | {c['grup_edat']:<6} | {c['codi_comportament']:<4} | {c['factor_codi']:<6} | {c['factor_edat']:<6} | {c['risc_final']:<6} | {c['segment']}")
```

Esta función procesa analíticamente los resultados finales para dar respuesta a las preguntas estratégicas de marketing. En lugar de evaluar de forma manual con estructuras condicionales, utiliza la función sorted() combinada con una clave lambda para generar un ranking descendente basado en el risc_final. De este modo, las respuestas de máximos (clientes prioritarios y rentables) y mínimos (perfiles descartables) se extraen directamente indexando las posiciones extremas ([0] y [-1]) de la colección ordenada.

```python
def mostrar_respostes(dades_processades):

    ranking_riesgo = sorted(dades_processades, key=lambda x: x["risc_final"], reverse=True)

    print("\n" + "="*50)
    print("      INFORME DE RESPUESTAS ESTRATÉGICAS")
    print("="*50)

    print(f"1. ¿A quién contactar primero?: A {ranking_riesgo[0]['nom']} (Riesgo: {ranking_riesgo[0]['risc_final']}) y {ranking_riesgo[1]['nom']} (Riesgo: {ranking_riesgo[1]['risc_final']}).")

    print("\n2. Diferencias por edad con mismo comportamiento:")
    print("   El factor de edad actúa como un ponderador de vulnerabilidad. Un cliente joven (18-25, factor 1.20)")
    print("   amplifica el riesgo de abandono. Por ejemplo, Laura Serra (23 años) y Mónica Pérez (47 años)")
    print("   tienen el mismo comportamiento (Código 4), pero Laura tiene un riesgo mayor (1.08) que Mónica (0.765)")
    print("   debido a que el público joven es estadísticamente más propenso a la volatilidad digital.")

    print(f"\n3. ¿A quién descartar por falta de interés?: A {ranking_riesgo[-1]['nom']} (Riesgo: {ranking_riesgo[-1]['risc_final']}), ya que solo navega y su interacción es mínima.")

    print(f"\n4. ¿Qué cliente sería el más rentable de recuperar?: {ranking_riesgo[0]['nom']}.")
    print(f"   Mensaje a enviarle: \"{ranking_riesgo[0]['missatge']}\"")
    print("="*50)
```

Actúa como el punto de entrada o función conductora (Main) del ejercicio. Su cometido es puramente organizativo: gestiona la inicialización del proceso de analítica invocando en estricto orden secuencial a la función de procesamiento de datos, a la de visualización tabular y, por último, a la encargada de emitir las conclusiones de negocio estratégicas.

```python
def main_marketing():

    dades_finals = processar_clients(clients)


    mostrar_diccionari(dades_finals)


    mostrar_respostes(dades_finals)


main_marketing()
```

En esta primera celda almacenamos la información en bruto aportada por el enunciado. Identificamos dos problemas de calidad de datos (data cleansing): los registros de los pacientes son cadenas de texto planas estructuradas de forma inconsistente con tuberías (|), y los síntomas presentan heterogeneidad en el uso de mayúsculas, espacios en blanco y separadores, difiriendo de las claves estandarizadas del diccionario médico nivells_risc.

```python
pacients = [
    "Maria|45|tos, Fiebre Alta, dolor_pecho",
    "Luis|33|dolor_cabeza,faTiga",
    "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
    "Jordi|52|fatiga , tos",
    "Anna|29|fiebre alta,dolor_cabeza"
]

nivells_risc = {
    "fiebre_alta": 3,
    "dificultad_respirar": 5,
    "dolor_pecho": 4,
    "tos": 1,
    "fatiga": 1,
    "dolor_cabeza": 1
}
```

Esta función es la encargada del saneamiento y homogeneización de cada síntoma individual. Recibe una cadena de texto y le aplica tres transformaciones en cadena: .lower() para eliminar la sensibilidad a las mayúsculas, .strip() para limpiar espacios en blanco innecesarios en los extremos, y .replace(" ", "_") para unificar los espacios internos con guiones bajos. Esto garantiza que el texto encaje perfectamente con las claves del diccionario de riesgos.

```python
def normalitzar_sintoma(sintoma_brut):
    sintoma_net = sintoma_brut.lower().strip().replace(" ", "_")
    return sintoma_net
```

El objetivo de esta función es calcular la gravedad cuantitativa del cuadro clínico del paciente. Recibe la lista de síntomas ya normalizados y, mediante un bucle for, acumula los valores numéricos correspondientes consultando el diccionario de referencia. Se utiliza el método seguro .get(sintoma, 0) para evitar que el programa se detenga con un error en caso de que se introduzca un síntoma no registrado en el sistema.

```python
def calcular_risc_pacient(llista_sintomes_nets):
    risc_total = 0
    for sintoma in llista_sintomes_nets:
        risc_total += nivells_risc.get(sintoma, 0)
    return risc_total
```

Esta función actúa como el transformador e integrador de registros (parser). Toma una línea de texto desestructurada y la fragmenta utilizando el método .split("|"). Convierte la edad a tipo entero, extrae la sublista de síntomas rompiéndola por comas, e itera sobre ellos aplicando la normalización y el cálculo de riesgo. Finalmente, empaqueta toda la información procesada en un diccionario limpio con campos explícitos.

```python
def processar_registre(cadena_pacient):
    parts = cadena_pacient.split("|")
    nom = parts[0]
    edat = int(parts[1])
    sintomes_bruts = parts[2].split(",")


    sintomes_nets = []
    for s in sintomes_bruts:
        sintomes_nets.append(normalitzar_sintoma(s))


    risc_total = calcular_risc_pacient(sintomes_nets)


    diccionari_pacient = {
        "nombre": nom,
        "edad": edat,
        "síntomas_normalizados": sintomes_nets,
        "riesgo_total": risc_total
    }
    return diccionari_pacient
```

Encargada de la capa de presentación de datos. Toma la colección de diccionarios médicos estructurados y los vuelca en la consola en formato tabular. Aplica técnicas de espaciado estático (:<N) para alinear perfectamente las columnas de texto y números, y emplea .join() para reconvertir las listas internas de síntomas en cadenas legibles y estéticas para el personal sanitario.

```python
def mostrar_tabla_pacientes(lista_procesada):
    print(f"{'NOMBRE':<10} | {'EDAD':<4} | {'RIESGO TOTAL':<12} | {'SÍNTOMAS NORMALIZADOS'}")
    print("-" * 75)
    for p in lista_procesada:

        sintomes_text = ", ".join(p["síntomas_normalizados"])
        print(f"{p['nombre']:<10} | {p['edad']:<4} | {p['riesgo_total']:<12} | {sintomes_text}")
```

Esta función implementa el protocolo de triaje médico requerido. Utiliza la función integrada sorted() junto con una función anónima lambda para ordenar el listado de pacientes de manera descendente según su riesgo_total. Posteriormente, mediante un bucle indexado con enumerate(..., 1), imprime el escalafón exacto de atención, garantizando que las emergencias críticas se sitúen prioritariamente en los primeros puestos.

```python
def mostrar_prioridades(lista_procesada):

    lista_ordenada = sorted(lista_procesada, key=lambda x: x["riesgo_total"], reverse=True)

    print("\n" + "="*55)
    print("      ORDEN DE ATENCIÓN URGENTE (TRIAJE MÉDICO)")
    print("="*55)

    for i, p in enumerate(lista_ordenada, 1):
        print(f"Prioridad {i}: {p['nombre']} (Edad: {p['edad']}) -> Riesgo Total: {p['riesgo_total']}")
    print("="*55)
```

Representa el cerebro o función principal del módulo (Main). Su responsabilidad es coordinar de principio a fin el flujo de ejecución del programa: gestiona el bucle de transformación de la lista de entrada, alimenta la base de datos temporal estructurada y acciona secuencialmente los módulos de impresión tabular e informes analíticos de urgencia.

```python
def main_medicina():

    pacientes_limpios = []

    for p_brut in pacients:
        paciente_estructurado = processar_registre(p_brut)
        pacientes_limpios.append(paciente_estructurado)


    print("--- DATASET MÉDICO PROCESADO ---")
    mostrar_tabla_pacientes(pacientes_limpios)

    mostrar_prioridades(pacientes_limpios)

main_medicina()
```

En esta celda se declaran las colecciones de datos iniciales. Siguiendo las directrices de optimización y la restricción de no utilizar condicionales if, almacenamos las métricas actuariales en el diccionario de diccionarios nivells_seguro. Este diseño permite mapear el código interno del cliente de manera directa ($O(1)$) con su categoría de riesgo, tarifa base y tasa de fraude estimada.

```python
clients = [
    {"nom": "Ana", "edat": 42, "codi": 0},
    {"nom": "Carlos", "edat": 30, "codi": 2},
    {"nom": "Isabel", "edat": 55, "codi": 3},
    {"nom": "Jorge", "edat": 40, "codi": 1},
    {"nom": "Marta", "edat": 28, "codi": 0}
]


nivells_seguro = {
    0: {"categoria": "Baix risc", "preu_base": 120, "fraude": 0.01},
    1: {"categoria": "Risc mitjà", "preu_base": 200, "fraude": 0.05},
    2: {"categoria": "Risc alt", "preu_base": 350, "fraude": 0.15},
    3: {"categoria": "Risc crític", "preu_base": 500, "fraude": 0.30}
}
```

Esta función realiza la clasificación e indexación del cliente sin lógica condicional. Utiliza el parámetro numérico codi como llave de acceso para extraer del diccionario global el bloque de datos técnicos del seguro, devolviendo una tupla con la categoría textual, el precio base y el porcentaje de fraude.

```python
def obtenir_parametres_segur(codi):
    info_segur = nivells_seguro[codi]
    return info_segur["categoria"], info_segur["preu_base"], info_segur["fraude"]
```

Función encargada del cálculo de la prima del seguro. Implementa la fórmula matemática provista por la compañía ($preu\_final = preu\_base \times (1 + edat/100)$), aplicando un recargo proporcional a la edad del tomador del seguro. El resultado se devuelve redondeado a dos decimales, respetando el formato monetario estándar.

```python
def calcular_preu_final(preu_base, edat):
    preu_final = preu_base * (1 + edat / 100)
    return round(preu_final, 2)
```

Esta función actúa como el motor de transformación de datos de la aplicación. Itera sobre el censo de clientes mediante un bucle for, coordina las subfunciones de extracción y cálculo matemático, y construye un nuevo set de datos unificado. Al encapsular la lógica por separado, garantizamos la modularidad y limpieza del flujo de datos.

```python
def processar_cartera_clients(llista_clients):
    clients_assegurats = []
    for c in llista_clients:
        cat, p_base, p_frau = obtenir_parametres_segur(c["codi"])
        p_final = calcular_preu_final(p_base, c["edat"])


        nou_assegurat = {
            "nombre": c["nom"],
            "edad": c["edat"],
            "categoría": cat,
            "precio_base": p_base,
            "precio_final": p_final,
            "probabilidad_frau": p_frau
        }
        clients_assegurats.append(nou_assegurat)
    return clients_assegurats
```

Esta función gestiona la salida de datos tabulares en consola. Utiliza alineación de texto estricta (:<N) para estructurar visualmente los registros simulando un DataFrame. Adicionalmente, aplica el formateo estadístico :.2% a la columna de fraude, transformando el valor decimal en un porcentaje limpio y legible para negocio (ej. 0.30 pasa a 30.00%).

```python
def mostrar_tabla_seguros(lista_procesada):
    print(f"{'NOMBRE':<10} | {'EDAD':<4} | {'CATEGORÍA':<12} | {'PRECIO BASE':<12} | {'PRECIO FINAL':<12} | {'PROB. FRAUDE'}")
    print("-" * 75)
    for c in lista_procesada:
        print(f"{c['nombre']:<10} | {c['edad']:<4} | {c['categoría']:<12} | {c['precio_base']:<12} | {c['precio_final']:<12} | {c['probabilidad_frau']:.2%}")
```

Módulo analítico que responde a los requerimientos de auditoría de la aseguradora. Emplea la función sorted() con expresiones lambda para crear dos clasificaciones independientes: una por riesgo reputacional/fraude y otra por volumen de facturación (precio final). Al extraer los elementos mediante indexación posicional, se automatiza el descubrimiento de los perfiles críticos sin recurrir a validaciones condicionales.

```python
def mostrar_analisi_risc(lista_procesada):

    ranking_fraude = sorted(lista_procesada, key=lambda x: x["probabilidad_frau"], reverse=True)
    ranking_precio = sorted(lista_procesada, key=lambda x: x["precio_final"], reverse=True)

    print("\n" + "="*60)
    print("         INFORME DE AUDITORÍA Y ANÁLISIS DE RIESGOS")
    print("="*60)

    print(f"» Cliente con más probabilidad de fraude: \n  {ranking_fraude[0]['nombre']} ({ranking_fraude[0]['probabilidad_frau']:.0%}) - Categoría: {ranking_fraude[0]['categoría']}.")

    print(f"\n» Cliente con la póliza (precio final) más alta: \n  {ranking_precio[0]['nombre']} ({ranking_precio[0]['precio_final']} €) - Edad: {ranking_precio[0]['edad']} años.")

    print(f"\n» Prioridad de revisión manual (Top 2 Alerta Fraude): \n  1. {ranking_fraude[0]['nombre']} (Riesgo Crítico) \n  2. {ranking_fraude[1]['nombre']} (Riesgo Alto)")
    print("="*60)
```

En esta celda preparamos los cimientos del generador. Importamos la librería numpy para gestionar la lógica de selección aleatoria avanzada mediante vectores. Además, empleamos el módulo nativo string para importar de manera estandarizada los repertorios de caracteres (mayúsculas, minúsculas, dígitos y signos de puntuación), evitando introducir errores tipográficos manuales.

```python
import numpy as np
import string


LETRAS_MIN = string.ascii_lowercase  # a-z
LETRAS_MAJ = string.ascii_uppercase  # A-Z
NUMEROS = string.digits              # 0-9
SIGNES = string.punctuation           # Símbolos especiales (!, @, #, $, etc.)
```

Esta función resuelve un requisito crítico de seguridad: asegurar que aparezca al menos un carácter de cada tipo seleccionado. Recibe los booleanos de configuración y, si están activos, utiliza np.random.choice() para preseleccionar inmediatamente un carácter obligatorio de ese grupo. Al mismo tiempo, va construyendo el "pool general" con todas las letras y números que el usuario ha permitido usar.

```python
def obtenir_caracters_disponibles(majuscules, minuscules, numeros, signes):
    pool_obligatori = []
    pool_general = ""


    if minuscules:
        pool_obligatori.append(np.random.choice(list(LETRAS_MIN)))
        pool_general += LETRAS_MIN
    if majuscules:
        pool_obligatori.append(np.random.choice(list(LETRAS_MAJ)))
        pool_general += LETRAS_MAJ
    if numeros:
        pool_obligatori.append(np.random.choice(list(NUMEROS)))
        pool_general += NUMEROS
    if signes:
        pool_obligatori.append(np.random.choice(list(SIGNES)))
        pool_general += SIGNES

    return pool_obligatori, list(pool_general)
```

Esta función se encarga del ensamblado y el cifrado posicional. Recibe los caracteres obligatorios, el conjunto general y la longitud solicitada. Rellena el espacio restante seleccionando elementos al azar con np.random.choice(..., size=N). Finalmente, ejecuta un algoritmo de desordenación masiva mediante np.random.shuffle(), rompiendo cualquier patrón predecible de ordenación y unificando el resultado en una cadena de texto única con .join().

```python
def completar_i_barrejar(pool_obligatori, pool_general, longitud):
    queden_per_omplir = longitud - len(pool_obligatori)


    caracters_extras = np.random.choice(pool_general, size=queden_per_omplir)


    contrasenya_llista = pool_obligatori + list(caracters_extras)


    np.random.shuffle(contrasenya_llista)


    return "".join(contrasenya_llista)
```

Es la función principal y el punto de acceso para el usuario. Cumple estrictamente con el requisito de diseño del enunciado al incorporar los parámetros nombrados con sus valores por defecto obligatorios (majuscules=True, signes=False, etc.). Además, incluye una salvaguarda lógica: si el usuario pide una longitud inferior al número de tipos de caracteres obligatorios seleccionados, reajusta dinámicamente la longitud para evitar fallos en el sistema.

```python
def generar_contrasenya(longitud=12, majuscules=True, minuscules=True, numeros=True, signes=False):
    obligatoris, generals = obtenir_caracters_disponibles(majuscules, minuscules, numeros, signes)


    if longitud < len(obligatoris):
        longitud = len(obligatoris)


    contrasenya_final = completar_i_barrejar(obligatoris, generals, longitud)
    return contrasenya_final
```

Esta función actúa como el entorno de pruebas (test bench) del departamento de Ciberseguridad. Invoca al generador bajo tres escenarios distintos para verificar la consistencia del algoritmo: valida el comportamiento de los valores por defecto del enunciado, pone a prueba la inserción de caracteres especiales de alta seguridad, y confirma la flexibilidad de la parametrización reducida.

```python
def main_ciberseguretat():
    print("==================================================")
    print("      TEST DE GENERACIÓN DE CONTRASENYAS SEGURAS")
    print("==================================================")


    pwd_defecte = generar_contrasenya(longitud=12)
    print(f"1. Configuración por defecto (Longitud 12): \n   -> {pwd_defecte}")


    pwd_maxima = generar_contrasenya(longitud=16, signes=True)
    print(f"\n2. Configuración Máxima Seguridad (Longitud 16 + Signos): \n   -> {pwd_maxima}")


    pwd_personalitzada = generar_contrasenya(longitud=8, majuscules=False, numeros=True)
    print(f"\n3. Configuración Personalizada (Longitud 8, solo min + num): \n   -> {pwd_personalitzada}")
    print("==================================================")


main_ciberseguretat()
```

Esta función realiza el parsing adaptado a la estructura real. En lugar de separar por comas, procesa cada línea aplicando .split("\t") para aislar los tres bloques: equipo local, el string del marcador y el equipo visitante. Posteriormente, un segundo despiece mediante .split("-") extrae las cifras de los goles, transformándolas a enteros (int) para poder operar matemáticamente.  

```python
def llegir_i_netejar_dades(ruta_arxiu):
    partits_nets = []
    with open(ruta_arxiu, "r", encoding="utf-8") as f:
        for linia in f:
            linia = linia.strip()
            if linia:
                eq_local, marcador, eq_visitant = linia.split("\t")


                gols_local, gols_visitant = marcador.split("-")

                partits_nets.append({
                    "local": eq_local.strip(),
                    "visitant": eq_visitant.strip(),
                    "gols_l": int(gols_local),
                    "gols_v": int(gols_visitant)
                })
    return partits_nets
```

Esta función mapea dinámicamente el catálogo de equipos únicos presentes en el histórico masivo. Inicializa un subdiccionario para cada club con tres contadores estadísticos en cero: goles a favor, goles en contra y puntos acumulados, previniendo excepciones por claves inexistentes durante la agregación.

```python
def inicialitzar_estadistiques(partits):
    estadistiques = {}
    for p in partits:
        if p["local"] not in estadistiques:
            estadistiques[p["local"]] = {"gols_marcats": 0, "gols_rebuts": 0, "punts": 0}
        if p["visitant"] not in estadistiques:
            estadistiques[p["visitant"]] = {"gols_marcats": 0, "gols_rebuts": 0, "punts": 0}
    return estadistiques
```

Representa el bloque de cálculo actuarial de la liga. Ejecuta un bucle sobre todo el histórico acumulando los datos de rendimiento ofensivo y defensivo de cada club. Al mismo tiempo, aplica las directrices reglamentarias para distribuir la puntuación de la clasificación general en base al signo del encuentro (3 puntos por ganar, 1 por empatar).

```python
def calcular_liga_completa(partits, estadistiques):
    for p in partits:
        loc, vis = p["local"], p["visitant"]
        gl, gv = p["gols_l"], p["gols_v"]


        estadistiques[loc]["gols_marcats"] += gl
        estadistiques[loc]["gols_rebuts"] += gv
        estadistiques[vis]["gols_marcats"] += gv
        estadistiques[vis]["gols_rebuts"] += gl


        if gl > gv:
            estadistiques[loc]["punts"] += 3
        elif gl < gv:
            estadistiques[vis]["punts"] += 3
        else:
            estadistiques[loc]["punts"] += 1
            text_vis = vis
            estadistiques[vis]["punts"] += 1

    return estadistiques
```

Transforma el diccionario estructurado en un ranking clasificatorio formal. Mediante la función sorted() y una tupla en la expresión lambda, establece un criterio de ordenación descendente robusto: prioriza los puntos y desempata por el volumen de goles a favor. Formatea los márgenes de impresión a un ancho fijo de 20 espacios para acomodar correctamente los nombres largos de los equipos.

```python
def mostrar_tabla_liga(estadistiques):
    llista_classificacio = []
    for equip, dades in estadistiques.items():
        llista_classificacio.append({
            "equipo": equip,
            "goles_marcados": dades["gols_marcats"],
            "goles_recibidos": dades["gols_rebuts"],
            "puntos": dades["punts"]
        })


    llista_ordenada = sorted(llista_classificacio, key=lambda x: (x["puntos"], x["goles_marcados"]), reverse=True)

    print(f"{'POS':<4} | {'EQUIPO':<20} | {'PUNTOS':<6} | {'GOLS MARCATS':<13} | {'GOLS REBUTS'}")
    print("-" * 65)
    for i, e in enumerate(llista_ordenada, 1):
        print(f"{i:<4} | {e['equipo']:<20} | {e['puntos']:<6} | {e['goles_marcados']:<13} | {e['goles_recibidos']}")

    return llista_ordenada
```

Genera las conclusiones finales requeridas por el enunciado en el idioma original solicitado. Utiliza la función max() con selectores lambda para extraer automáticamente los dos extremos de la estadística de juego: la delantera más demoledora (máximos goles a favor) y la defensa más castigada de la liga (máximos goles en contra, corrigiendo la traducción inexacta de la plantilla anterior).

```python
def mostrar_informe_deportivo(lista_ordenada):
    max_goleador = max(lista_ordenada, key=lambda x: x["goles_marcados"])

    max_golejat = max(lista_ordenada, key=lambda x: x["goles_recibidos"])

    print("\n" + "="*55)
    print("      INFORME DE ESTADÍSTICAS FINALES (DATOS REALES)")
    print("="*55)
    print(f"Equip més golejador: {max_goleador['equipo']} ({max_goleador['goles_marcados']} gols)")
    print(f"Equip més golejat:   {max_golejat['equipo']} ({max_golejat['goles_recibidos']} gols)")

    print("\nClassificació final:")
    for i, e in enumerate(lista_ordenada, 1):
        print(f"{i}. {e['equipo']}")
    print("="*55)
```

La función principal orquesta y ejecuta de manera secuencial los bloques de código del programa. Centraliza el flujo controlando la entrada del archivo de texto plano real, la transformación de datos y la salida limpia por consola de los entregables que solicitó el departamento analítico deportivo.

```python
def main_deportes():
    ruta = "historic partits.txt"


    partidos = llegir_i_netejar_dades(ruta)


    diccionari_vuit = inicialitzar_estadistiques(partidos)


    diccionari_calculat = calcular_liga_completa(partidos, diccionari_vuit)


    print("--- CLASIFICACIÓN GENERAL DE LA LIGA CATALANA ---")
    lista_ranking = mostrar_tabla_liga(diccionari_calculat)


    mostrar_informe_deportivo(lista_ranking)


main_deportes()
```
